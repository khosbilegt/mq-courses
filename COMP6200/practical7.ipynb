```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
print("notebook is working")
# File
data = pd.read_excel("practical7.xlsx", sheet_name="Sheet1")
# Select Columns
coords = data[["Latitude", "Longitude"]]
# DBSCAN
model = DBSCAN(eps=0.5, min_samples=2)
labels = model.fit_predict(coords)
# Scatter Plot
data['cluster'] = labels
data.plot.scatter(
     x="Longitude",
     y="Latitude",
     c=labels,
     colormap="tab10"
)
```

```python
score = silhouette_score(coords, labels)
print(f"Silhouette score: {score:.2f}")
```

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data.plot.scatter(x='sepal length (cm)', y='sepal width (cm)')
```

```python
import sklearn.cluster
kmeans = sklearn.cluster.KMeans(n_clusters=3, random_state=3)

kmeans_cluster = kmeans.fit_predict(data[iris.feature_names])

score = sklearn.metrics.silhouette_score(
     data[iris.feature_names],
     kmeans_cluster
)

kmeans_data = kmeans_cluster
kmeans_data['kmeans_cluster'] = kmeans_cluster
kmeans_data.plot.scatter(x='sepal length (cm)', y='sepal width (cm)')

```
